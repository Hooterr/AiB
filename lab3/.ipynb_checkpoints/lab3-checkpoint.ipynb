```python
# Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.

df = pd.read_sql('''
select 
    Count(distinct country.country)
from 
    customer
join address ON customer.address_id = address.address_id
join city on address.city_id = city.city_id
join country on city.city_id = country.country_id
''',con=connection)
df
```

```python
# Gdzie mieszkają członkowie personelu?

df = pd.read_sql('''
select 
    address
from 
    staff
join address ON staff.address_id = address.address_id

''',con=connection)
df
```

```python
# Ilu pracowników mieszka w Argentynie lub Hiszpanii?

df = pd.read_sql('''
select 
   COUNT(country.country_id)
from 
    staff
join address ON staff.address_id = address.address_id
join city on address.city_id = city.city_id
join country on city.country_id = country.country_id
WHERE country.country = 'Argentina' or country.country = 'Spain'
''',con=connection)
df
```

```python
# Jakie kategorie filmów zostały wypożyczone przez klientów?

df = pd.read_sql('''
select 
    DISTINCT category.name
from 
    rental
join inventory ON rental.inventory_id = inventory.inventory_id
join film ON inventory.film_id = film.film_id
join film_category ON film_category.film_id = film.film_id
join category ON category.category_id = film_category.category_id
''',con=connection)
df
```

```python
# Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

df = pd.read_sql('''
select 
     DISTINCT category.name
from 
    rental
join inventory ON rental.inventory_id = inventory.inventory_id
join film ON inventory.film_id = film.film_id
join film_category ON film_category.film_id = film.film_id
join category ON category.category_id = film_category.category_id
where rental.rental_id In (SELECT 
    rental.rental_id
    from rental
        join customer ON customer.customer_id = rental.customer_id
        join address ON customer.address_id = address.address_id
        join city ON address.city_id = city.city_id
        join country ON city.country_id = country.country_id
        where country.country = 'Spain')
''',con=connection)
df
```
